```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Step 1: Reading the CSV file and extracting features and labels
x = pd.read_csv("C:/Users/ASUS/Downloads/Machine Learning Full/SVM/animal_data.csv")
```

```python
# Extract the features (height, weight)
x = x[['height', 'weight']].values

# Extract the class labels (assuming they are in the second column)
y = x[:, 1]  # Assuming the class labels are in the second column (0 and 1)
```

```python
# Step 2: Convert continuous labels into discrete binary labels based on a threshold
threshold = 150  # Define the threshold
y_discrete = np.where(y > threshold, 1, 0)  # Convert to 0 or 1 based on the threshold
```

```python
# Display the shape of x and y
print(x.shape)  # Should print (569, 2) if there are 569 samples and 2 features
print(x)  # Features
print(y_discrete)  # Binary class labels
```

```python

# Step 3: Split the data into training and test sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x, y_discrete, test_size=0.2, random_state=42)
```

```python
# Step 4: Train the classifier
clf = SVC(kernel='linear')  # Create an SVC classifier with a linear kernel
clf.fit(x_train, y_train)  # Fit the model to the training data
```

```python
# Step 5: Make predictions on the test set
y_pred = clf.predict(x_test)
```

```python
# Step 6: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)  # Calculate accuracy
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
# Step 7: Predicting new data
new_data = np.array([[5.5, 150], [9.0, 180]])  # New data (height, weight)
predictions = clf.predict(new_data)
print("Predictions for new data:", predictions)
```
